```python
import numpy as np
import pandas as pd

def step1_normal_vs_unbiased(score, step1_emd_df, step1_eemd_df, step1_ceemdan_df):
    step1_emd_ss = np.round((step1_emd_df['Score_x']/step1_emd_df['Score_y'] - 1)*100,2)
    step1_emd_ss.name ='Normal vs Unbiased (%)'
    
    step1_eemd_ss = np.round((step1_eemd_df['Score_x']/step1_eemd_df['Score_y'] - 1)*100,2)
    step1_eemd_ss.name ='Normal vs Unbiased (%)'
    
    step1_ceemdan_ss = np.round((step1_ceemdan_df['Score_x']/step1_ceemdan_df['Score_y'] - 1)*100,2)
    step1_ceemdan_ss.name ='Normal vs Unbiased (%)'
    
    
    df = pd.DataFrame()
    df.insert(loc=0, column='EMD_{0} Normal vs Unbiased (%)'.format(score.title()),
              value=step1_emd_ss)
    df.insert(loc=1, column='EEMD_{0} Normal vs Unbiased (%)'.format(score.title()),
              value=step1_eemd_ss)
    df.insert(loc=2, column='CEEMDAN_{0} Normal vs Unbiased (%)'.format(score.title()),
          value=step1_ceemdan_ss)
    return df

#score = 'acc' or score = 'f1'
#step1_merge_step2('acc', step1_emd_acc_df, step2_emd_acc_df)
def step1_merge_step2(score, step1_df, step2_df):
    
    # selecting series from step1_df
    s1_normal_score = step1_df['Score_x'].tolist()
    s1_unbiased_score = step1_df['Score_y'].tolist()

    # selecting series from step2_df
    s2_normal_score = step2_df['Normal_Nested_{0}'.format(score.title())].tolist()
    s2_unbiased_score = step2_df['Unbiased_Nested_{0}'.format(score.title())].tolist()

    # creating the merged dataframe
    ss_index = ['KNN / Grid_KNN', 'RandomForest / Grid_RandomForest','SVM / Grid_SVM']
    df_index = ['{0}_{1}_{2}'.format('Step1', 'Normal', score.title()),
                '{0}_{1}_{2}'.format('Step2', 'Normal', score.title()),
                '{0}_{1}_{2}'.format('Step1', 'Unbiased', score.title()),
                '{0}_{1}_{2}'.format('Step2', 'Unbiased', score.title())]
    s1s2_df = pd.DataFrame([pd.Series(s1_normal_score, index=ss_index),
                            pd.Series(s2_normal_score, index=ss_index),
                            pd.Series(s1_unbiased_score, index=ss_index),
                            pd.Series(s2_unbiased_score, index=ss_index)], index=df_index).round(4)
    
    return s1s2_df.T

def merge_all_steps(score, step1_df, step2_df, step3_df):
   
    # selecting the series from step3_df
    s3_normal_score = step3_df['Normal_Nested_{0}'.format(score.title())].tolist()
    s3_unbaised_score = step3_df['Unbiased_Nested_{0}'.format(score.title())].tolist()
    # 
    s1s2_df = step1_merge_step2(score, step1_df, step2_df)
    #df.insert(loc=idx, column='A', value=new_col)
    s1s2_df.insert(loc=2, column='Step3_Normal_{0}'.format(score.title()), value=s3_normal_score)
    s1s2_df.insert(loc=5, column='Step3_Unbiased_{0}'.format(score.title()), value=s3_unbaised_score)
    return s1s2_df.round(4)

#step1_merge_step2('acc', step1_emd_acc_df, step2_emd_acc_df)
def step23_normal_vs_unbiased(score,step23_emd_df, step23_eemd_df, step23_ceemdan_df):
    # selecting series from step23_df
    emd_normal_ss= step23_emd_df['Normal_Nested_{0}'.format(score.title())]
    emd_unbaised_ss = step23_emd_df['Unbiased_Nested_{0}'.format(score.title())]
    step23_emd_ss = np.round((emd_normal_ss/emd_unbaised_ss-1)*100,2)
    #step2_emd_ss.name ='Normal vs Unbiased (%)'
    #
    eemd_normal_ss = step23_eemd_df['Normal_Nested_{0}'.format(score.title())]
    eemd_unbaised_ss = step23_eemd_df['Unbiased_Nested_{0}'.format(score.title())]
    step23_eemd_ss = np.round((eemd_normal_ss/eemd_unbaised_ss-1)*100,2)
    #step2_eemd_ss.name ='Normal vs Unbiased (%)'
    #
    ceemdan_normal_ss = step23_ceemdan_df['Normal_Nested_{0}'.format(score.title())]
    ceemdan_unbaised_ss = step23_ceemdan_df['Unbiased_Nested_{0}'.format(score.title())]
    step23_ceemdan_ss = np.round((ceemdan_normal_ss/ceemdan_unbaised_ss-1)*100,2)
    #step2_ceemdan_ss.name ='Normal vs Unbiased (%)'
    
    df = pd.DataFrame()
    df.insert(loc=0, column='EMD_{0} Normal / Unbiased (%)'.format(score.title()),
              value=step23_emd_ss)
    df.insert(loc=1, column='EEMD_{0} Normal / Unbiased (%)'.format(score.title()),
              value=step23_eemd_ss)
    df.insert(loc=2, column='CEEMDAN_{0} Normal / Unbiased (%)'.format(score.title()),
          value=step23_ceemdan_ss)
    return df
```

```python
def step2_vs_step1(score,step1_emd_df,step1_eemd_df, step1_ceemdan_df,
                   step2_emd_df, step2_eemd_df, step2_ceemdan_df):
  
    s2s1_emd_normal= step2_emd_df['Normal_Nested_{0}'.format(score.title())].values \
                     / step1_emd_df['Score_x'].values
    s2s1_eemd_normal= step2_eemd_df['Normal_Nested_{0}'.format(score.title())].values \
                      / step1_eemd_df['Score_x'].values
    s2s1_ceemdan_normal= step2_ceemdan_df['Normal_Nested_{0}'.format(score.title())].values \
                        / step1_ceemdan_df['Score_x'].values
    
    s2s1_emd_unbiased= step2_emd_df['Unbiased_Nested_{0}'.format(score.title())].values \
                        / step1_emd_df['Score_y'].values
    s2s1_eemd_unbiased= step2_eemd_df['Unbiased_Nested_{0}'.format(score.title())].values \
                        / step1_eemd_df['Score_y'].values
    s2s1_ceemdan_unbiased= step2_ceemdan_df['Unbiased_Nested_{0}'.format(score.title())].values \
                        / step1_ceemdan_df['Score_y'].values
    
    
    # creating the merged dataframe
    ss_index = ['KNN / Grid_KNN', 'RandomForest / Grid_RandomForest','SVM / Grid_SVM']
    df_index = ['{0} {1}_{2} {3} {4}'.format('Step2 / Step1', 'EMD', score.title(),'Normal','(%)'), 
                '{0} {1}_{2} {3} {4}'.format('Step2 / Step1', 'EEMD', score.title(),'Normal','(%)'),
                '{0} {1}_{2} {3} {4}'.format('Step2 / Step1', 'CEEMDAN', score.title(),'Normal','(%)'),
                '{0} {1}_{2} {3} {4}'.format('Step2 / Step1', 'EMD', score.title(),'Unbiased','(%)'), 
                '{0} {1}_{2} {3} {4}'.format('Step2 / Step1', 'EEMD', score.title(),'Unbiased','(%)'),
                '{0} {1}_{2} {3} {4}'.format('Step2 / Step1', 'CEEMDAN', score.title(),'Unbiased','(%)')]
                #EMD_Acc Normal
    s2s1_df = pd.DataFrame([pd.Series((s2s1_emd_normal-1)*100, index=ss_index),
                            pd.Series((s2s1_eemd_normal-1)*100, index=ss_index),
                            pd.Series((s2s1_ceemdan_normal-1)*100, index=ss_index),
                            pd.Series((s2s1_emd_unbiased-1)*100, index=ss_index),
                            pd.Series((s2s1_eemd_unbiased-1)*100, index=ss_index),
                            pd.Series((s2s1_ceemdan_unbiased-1)*100, index=ss_index)],
                           index=df_index).round(2)
    return s2s1_df.T

def step3_vs_step2(score,step2_emd_df,step2_eemd_df, step2_ceemdan_df,
                   step3_emd_df, step3_eemd_df, step3_ceemdan_df):
    #
    s3s2_emd_normal= step3_emd_df['Normal_Nested_{0}'.format(score.title())].values \
                        / step2_emd_df['Normal_Nested_{0}'.format(score.title())].values
    s3s2_eemd_normal= step3_eemd_df['Normal_Nested_{0}'.format(score.title())].values \
                        / step2_eemd_df['Normal_Nested_{0}'.format(score.title())].values
    s3s2_ceemdan_normal= step3_ceemdan_df['Normal_Nested_{0}'.format(score.title())].values \
                        / step2_ceemdan_df['Normal_Nested_{0}'.format(score.title())].values
    #    
    s3s2_emd_unbiased= step3_emd_df['Unbiased_Nested_{0}'.format(score.title())].values \
                        / step2_emd_df['Unbiased_Nested_{0}'.format(score.title())].values
    s3s2_eemd_unbiased= step3_eemd_df['Unbiased_Nested_{0}'.format(score.title())].values \
                        / step2_eemd_df['Unbiased_Nested_{0}'.format(score.title())].values
    s3s2_ceemdan_unbiased= step3_ceemdan_df['Unbiased_Nested_{0}'.format(score.title())].values \
                        / step2_ceemdan_df['Unbiased_Nested_{0}'.format(score.title())].values
    
    
    # creating the merged dataframe
    ss_index = ['KNN / Grid_KNN', 'RandomForest / Grid_RandomForest','SVM / Grid_SVM']
    df_index = ['{0} {1}_{2} {3} {4}'.format('Step3 / Step2', 'EMD', score.title(),'Normal','(%)'), 
                '{0} {1}_{2} {3} {4}'.format('Step3 / Step2', 'EEMD', score.title(),'Normal','(%)'),
                '{0} {1}_{2} {3} {4}'.format('Step3 / Step2', 'CEEMDAN', score.title(),'Normal','(%)'),
                '{0} {1}_{2} {3} {4}'.format('Step3 / Step2', 'EMD', score.title(),'Unbiased','(%)'), 
                '{0} {1}_{2} {3} {4}'.format('Step3 / Step2', 'EEMD', score.title(),'Unbiased','(%)'),
                '{0} {1}_{2} {3} {4}'.format('Step3 / Step2', 'CEEMDAN', score.title(),'Unbiased','(%)')]
                #EMD_Acc Normal
    s3s2_df = pd.DataFrame([pd.Series((s3s2_emd_normal-1)*100, index=ss_index),
                            pd.Series((s3s2_eemd_normal-1)*100, index=ss_index),
                            pd.Series((s3s2_ceemdan_normal-1)*100, index=ss_index),
                            pd.Series((s3s2_emd_unbiased-1)*100, index=ss_index),
                            pd.Series((s3s2_eemd_unbiased-1)*100, index=ss_index),
                            pd.Series((s3s2_ceemdan_unbiased-1)*100, index=ss_index)],
                           index=df_index).round(2)
    return s3s2_df.T
```
